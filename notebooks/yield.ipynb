### O que  o Yield?
O Yield no contexto de imveis ou em investimentos em geral refere-se ao retorno financeiro que um imvel gera em relao ao valor de compra. No caso,  o clculo da rentabilidade do aluguel em relao ao preo de venda. A frmula bsica do Yield :

$$
\text{Yield} = \frac{\text{Aluguel Anual}}{\text{Preo de Venda}}
$$

- **Aluguel Anual**: o valor do aluguel mensal multiplicado por 12 (para obter o valor anual).
- **Preo de Venda**: o preo de venda do imvel.

Resumindo o Yield ajuda a determinar o quo rentvel  um investimento em aluguel em comparao com o preo do imvel.

- **Eixo Y**: Representa o Yield, ou seja, o retorno do aluguel em relao ao preo de venda do imvel.
- **Linha Verde**: Mostra a variao do Yield ao longo do tempo, permitindo ao investidor visualizar se o retorno sobre o aluguel est aumentando ou diminuindo.
