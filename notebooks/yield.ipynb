### O que é o Yield?
O Yield no contexto de imóveis ou em investimentos em geral refere-se ao retorno financeiro que um imóvel gera em relação ao valor de compra. No caso, é o cálculo da rentabilidade do aluguel em relação ao preço de venda. A fórmula básica do Yield é:

$$
\text{Yield} = \frac{\text{Aluguel Anual}}{\text{Preço de Venda}}
$$

- **Aluguel Anual**: o valor do aluguel mensal multiplicado por 12 (para obter o valor anual).
- **Preço de Venda**: o preço de venda do imóvel.

Resumindo o Yield ajuda a determinar o quão rentável é um investimento em aluguel em comparação com o preço do imóvel.
