## 📊 Interpretação da Correlação de Pearson

O coeficiente de **correlação de Pearson** é uma medida estatística que indica a força e a direção da relação linear entre duas variáveis. Ele varia de **-1** a **+1**, e é amplamente utilizado para entender como duas variáveis estão relacionadas entre si, o que pode ser muito útil na análise de dados do mercado imobiliário.

### 🎯 Objetivo da análise

Calcular a **correlação de Pearson** entre duas variáveis permite entender se e como essas variáveis estão relacionadas. O objetivo principal aqui é:

- **Verificar a relação entre variáveis**, como, por exemplo, a relação entre **IPCA (inflação)** e **Preço do metro quadrado**.
- Avaliar a **força da relação linear** entre as variáveis, indicando se um aumento ou diminuição em uma variável pode ser acompanhado por um aumento ou diminuição na outra.
- **Escolher variáveis relevantes** para análises mais profundas, como modelos de previsão de preços ou análise de fatores econômicos.

### 📌 Como interpretar o coeficiente de correlação de Pearson

O valor da correlação de Pearson varia entre **-1 e +1** e é interpretado da seguinte forma:

- **+1** → **Correlação positiva perfeita**: à medida que uma variável aumenta, a outra também aumenta de forma perfeitamente linear.
- **-1** → **Correlação negativa perfeita**: à medida que uma variável aumenta, a outra diminui de forma perfeitamente linear.
- **0** → **Sem correlação linear**: as duas variáveis não têm relação linear.
- **0.1 a 0.3** → **Correlação fraca positiva**.
- **0.3 a 0.5** → **Correlação moderada positiva**.
- **0.5 a 0.7** → **Correlação forte positiva**.
- **0.7 a 0.9** → **Correlação muito forte positiva**.
- **-0.1 a -0.3** → **Correlação fraca negativa**.
- **-0.3 a -0.5** → **Correlação moderada negativa**.
- **-0.5 a -0.7** → **Correlação forte negativa**.
- **-0.7 a -0.9** → **Correlação muito forte negativa**.

### 🧐 O que observar

- **Correlação positiva**: Se uma variável tem uma forte correlação positiva com o preço de venda ou aluguel de imóveis, isso sugere que, à medida que a variável aumenta, o preço tende a aumentar também. Exemplos podem incluir indicadores econômicos como o **PIB** ou a **taxa de juros**.
- **Correlação negativa**: Uma forte correlação negativa pode indicar que, à medida que uma variável aumenta, o preço de venda ou aluguel tende a diminuir. Isso pode ser observado em variáveis como **taxas de juros** ou **oferta de imóveis**.
- **Correlação próxima de 0**: Se o coeficiente de correlação for próximo de 0, isso indica que não há uma relação linear clara entre as duas variáveis. Isso pode ser importante para descartar variáveis irrelevantes em análises futuras.
- **Correlação muito alta entre variáveis**: Se duas variáveis apresentam uma correlação muito alta (acima de 0.9 ou abaixo de -0.9), pode ser que elas sejam redundantes, ou seja, fornecem informações semelhantes. Nesse caso, você pode querer considerar uma dessas variáveis para evitar multicolinearidade em modelos preditivos.

### 💡 Dica para a análise

Embora a correlação de Pearson seja uma ferramenta útil para identificar relações lineares, é importante lembrar que **correlação não implica causalidade**. Portanto, ao analisar as correlações, procure sempre fazer uma interpretação cuidadosa e considerar outras variáveis ou análises complementares.

Uma boa prática é usar o **coeficiente de correlação de Pearson como uma primeira abordagem** para entender a relação entre as variáveis, mas depois investigar mais a fundo as possíveis causas ou fatores que podem estar influenciando essas relações.
