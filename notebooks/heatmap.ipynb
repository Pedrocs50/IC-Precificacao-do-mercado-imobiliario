```python
import sys
import os

sys.path.append(os.path.abspath("../src"))

from heatmap import DataLoader, CorrelationCalculator, HeatmapGenerator

caminho_dados = "../data/processed/Dados-Unificados.csv"
caminho_venda_aluguel = "../data/processed/Preço-venda-aluguel.csv"

loader = DataLoader(caminho_dados, caminho_venda_aluguel)
df_completo = loader.combinar_dados()
df_numerico = loader.selecionar_colunas_numericas(df_completo)

calc = CorrelationCalculator(df_numerico)
correlacao = calc.calcular_correlacao(metodo='spearman')

heatmap = HeatmapGenerator(correlacao)
heatmap.gerar_heatmap()
```

## 📊 Interpretação do Heatmap de Correlação

O heatmap gerado mostra a **correlação estatística entre variáveis numéricas** relevantes para o mercado imobiliário, incluindo indicadores econômicos e dados de venda/aluguel de imóveis.

### 🎯 Objetivo da análise

Identificar **relações significativas** entre variáveis que possam influenciar o comportamento dos preços de venda e aluguel, com foco em:

- Quais variáveis estão **positivamente correlacionadas** com os preços de venda/aluguel?
- Quais têm **correlação negativa**, podendo indicar efeito contrário?
- Existem variáveis **altamente correlacionadas entre si**, o que pode indicar redundância?

### 📌 Como interpretar os valores

- Correlação próxima de **+1.0** → relação positiva forte (ambas sobem ou descem juntas).
- Correlação próxima de **-1.0** → relação negativa forte (uma sobe, a outra desce).
- Correlação próxima de **0.0** → sem relação linear ou monotônica evidente.

### 🧐 O que observar

- **Indicadores macroeconômicos (como inflação, taxa de juros, etc)** que tenham alta correlação com os preços de venda ou aluguel — podem ser variáveis preditoras importantes.
- **Relação entre os diferentes defasagens (1D, 2D, etc)** — essas colunas representam o comportamento dos preços ao longo do tempo e ajudam a entender a persistência ou mudança nas tendências.
- Variáveis com **correlação muito alta entre si** (> 0.9 ou < -0.9) podem ser redundantes — importante considerar em análises futuras (como regressões ou modelos preditivos).

### 💡 Dica para a análise

Use o heatmap como um **ponto de partida exploratório**. A correlação não implica causalidade, mas ajuda a filtrar quais variáveis merecem investigação mais profunda. Uma boa prática é cruzar essa análise com conhecimento do mercado ou modelos estatísticos.
