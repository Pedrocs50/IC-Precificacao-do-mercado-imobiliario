```python
import sys
import os

sys.path.append(os.path.abspath("../src"))

from heatmap import DataLoader, CorrelationCalculator, HeatmapGenerator

caminho_dados = "../data/processed/Dados-Unificados.csv"
caminho_venda_aluguel = "../data/processed/Preo-venda-aluguel.csv"

loader = DataLoader(caminho_dados, caminho_venda_aluguel)
df_completo = loader.combinar_dados()
df_numerico = loader.selecionar_colunas_numericas(df_completo)

calc = CorrelationCalculator(df_numerico)
correlacao = calc.calcular_correlacao(metodo='spearman')

heatmap = HeatmapGenerator(correlacao)
heatmap.gerar_heatmap()
```

##  Interpretao do Heatmap de Correlao

O heatmap gerado mostra a **correlao estatstica entre variveis numricas** relevantes para o mercado imobilirio, incluindo indicadores econmicos e dados de venda/aluguel de imveis.

###  Objetivo da anlise

Identificar **relaes significativas** entre variveis que possam influenciar o comportamento dos preos de venda e aluguel, com foco em:

- Quais variveis esto **positivamente correlacionadas** com os preos de venda/aluguel?
- Quais tm **correlao negativa**, podendo indicar efeito contrrio?
- Existem variveis **altamente correlacionadas entre si**, o que pode indicar redundncia?

###  Como interpretar os valores

- Correlao prxima de **+1.0**  relao positiva forte (ambas sobem ou descem juntas).
- Correlao prxima de **-1.0**  relao negativa forte (uma sobe, a outra desce).
- Correlao prxima de **0.0**  sem relao linear ou monotnica evidente.

###  O que observar

- **Indicadores macroeconmicos (como inflao, taxa de juros, etc)** que tenham alta correlao com os preos de venda ou aluguel  podem ser variveis preditoras importantes.
- **Relao entre os diferentes defasagens (1D, 2D, etc)**  essas colunas representam o comportamento dos preos ao longo do tempo e ajudam a entender a persistncia ou mudana nas tendncias.
- Variveis com **correlao muito alta entre si** (> 0.9 ou < -0.9) podem ser redundantes  importante considerar em anlises futuras (como regresses ou modelos preditivos).

###  Dica para a anlise

Use o heatmap como um **ponto de partida exploratrio**. A correlao no implica causalidade, mas ajuda a filtrar quais variveis merecem investigao mais profunda. Uma boa prtica  cruzar essa anlise com conhecimento do mercado ou modelos estatsticos.
